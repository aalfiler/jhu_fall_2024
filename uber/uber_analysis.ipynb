```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import seaborn as sns
```

```python
# Read the Excel file
try:
    df = pd.read_excel('uber_data.xlsx', index_col=0)

except FileNotFoundError:
    print("Error: Excel file not found in the current directory")
except Exception as e:
    print(f"An error occurred: {str(e)}")
```

```python
#Create Word Cloud from Keywords column
plt.figure(figsize=(10, 6))
wordcloud = WordCloud(
    width=800,
    height=400,
    background_color='white'
).generate(' '.join(df['Keywords'].dropna().astype(str)))

plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud of Keywords')
plt.show()
```

```python
# Create bar graph for Sentiment column
plt.figure(figsize=(10, 6))
sentiment_counts = df['Sentiment'].value_counts()
sns.barplot(x=sentiment_counts.index, y=sentiment_counts.values)
plt.title('Distribution of Sentiments')
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Create bar graph for Rider/Driver column
plt.figure(figsize=(10, 6))
rider_driver_counts = df['Rider/Driver'].value_counts()
sns.barplot(x=rider_driver_counts.index, y=rider_driver_counts.values)
plt.title('Distribution of Rider vs Driver Comments')
plt.xlabel('Category')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

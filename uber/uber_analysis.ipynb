```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import seaborn as sns
```

```python
# Read the Excel file
try:
    df = pd.read_excel('uber_data.xlsx', index_col=0)

except FileNotFoundError:
    print("Error: Excel file not found in the current directory")
except Exception as e:
    print(f"An error occurred: {str(e)}")
```

```python
#Create Word Cloud from Keywords column
plt.figure(figsize=(10, 6))
wordcloud = WordCloud(
    width=800,
    height=400,
    background_color='white'
).generate(' '.join(df['Keywords'].dropna().astype(str)))

plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud of Keywords')
plt.show()

plt.savefig('wordcloud.png')
plt.close()
```

```python
#Create grouped bar graph for Sentiment by Rider/Driver
plt.figure(figsize=(7, 5))

# Create a cross-tabulation of Rider/Driver and Sentiment
sentiment_by_category = pd.crosstab(df['Rider/Driver'], df['Sentiment'])

# Create the grouped bar chart
sentiment_by_category.plot(kind='bar', ax=plt.gca())
plt.title('Distribution of Sentiments')
plt.xlabel('Category')
plt.ylabel('Count')
plt.legend(title='Sentiment')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

plt.savefig('distributionSentiment.png')
plt.close()
```
